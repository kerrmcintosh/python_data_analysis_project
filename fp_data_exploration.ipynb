```python
expl_follow_ups = jigger_details.drop(jigger_details.columns.difference(['id', 'past_week_30_initial', 'past_week_30_final',
       'currentlyRegularActive', 'currentlyRegularActiveFinal', 'simd_d', 'joinDate', 'followup_received_date']), 1)
```

```python
import datetime
from datetime import datetime
```

```python
expl_follow_ups['year'] = pd.to_datetime(expl_follow_ups.joinDate, errors ="coerce" )
expl_follow_ups['year'] = expl_follow_ups['year'].dt.year
```

```python
expl_follow_ups = expl_follow_ups[expl_follow_ups['year'] >= 2017]
```

```python
# jiggers with activity levels before and after  and then by simd
```

```python
# 8,883 jiggers 2017 onwards
expl_follow_ups.shape
```

```python
# get rid of simd 0 
simd_follow_ups = expl_follow_ups[expl_follow_ups['simd_d'] > 0]
simd_follow_ups.shape
```

```python
#1,480 walkers in simd 1-3 since 2017
simd_follow_ups = simd_follow_ups[simd_follow_ups['simd_d'] < 4]
simd_follow_ups.shape
```

```python
#simd_follow_ups.to_csv('codeclan/simd_follow_ups.csv', index = False, header=True)
# csv taken for inspection
```

```python
simd_by_year = simd_follow_ups.groupby(['simd_d', 'year']).agg({'id' : len}).round(2)
```

```python
# jiggers in z1-3 since 2017
simd_by_year
```

```python
simd_follow_up_actlevel.groupby(['simd_d', 'year']).agg({'id' : len}).round(2)
```
